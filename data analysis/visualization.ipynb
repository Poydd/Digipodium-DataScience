```python
x = np.arange(-5, 5, 0.2)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(15,5))
axs[0].plot(x, y1, 'r--')
axs[1].plot(x, y2, 'b--')
axs[2].plot(x, y3,'g--')
axs[0].set_title('sin(x)')
axs[1].set_title('cos(x)')
axs[2].set_title('tan(x)')
plt.show()
```

implicit vs explicit graphs

```python
# plt  -> implicit figure and axes creation
x = np.linspace(0, 10, 25)
plt.plot(x, x, '-', label='linear')
plt.plot(x, x**2, '--', label='quadratic')
plt.plot(x, x**3, 'o-', label='cubic')
plt.xlabel('my x label')
plt.ylabel('my y label')
plt.title('simple plot with implicit graph setting')
plt.legend()
plt.show()
```

```python
x = np.linspace(-10, 10, 25)
plt.plot(x, x, '-', label='linear')
plt.plot(x, x**2, '--', label='quadratic')
plt.plot(x, x**3, 'o-', label='cubic')
plt.xlabel('my x label')
plt.ylabel('my y label')
plt.title('simple plot with implicit graph setting')
plt.legend()
plt.show()
```

```python
# explicit
x = np.linspace(-10, 10, 25)
fig, ax = plt.subplots()   # fig and ax are objects
plt.plot(x, x, 'g-', label='linear')
plt.plot(x, x**2, 'r--', label='quadratic')
plt.plot(x, x**3, 'o:', label='quadratic')
ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.set_title('simple plot')
plt.legend()
plt.show()
```

```python
# using dictionary for graph 
data = {
    "a":np.arange(50),
    "b":np.random.randint(0, 50, 50),
    "c":np.random.randn(50),
    "d":np.arange(50) + 10 * np.random.randn(50)
}
data['d'] = np.abs(data['d']) * 10

fig, ax = plt.subplots(figsize=(10,3))
ax.scatter('a', 'b', c='c', s='d', data=data)
ax.set_xlabel('entry a')
ax.set_ylabel('entry b')
ax.set_title('scatter plot')
plt.show()
```

```python
data = {
    'names' : ['taylor', 'selena', 'kai', 'zee'],
   'english': [78, 68, 97, 89],
   'math': [79,98,99,94],
   'hindi':[94,96,99,99],
}
```

```python
def my_custom_plotter(ax, x_value, y_value, param_dict):
    out = ax.bar(x_value, y_value, **param_dict)
    return out
```

```python
fig, axs = plt.subplots(1, 3, figsize=(15,3))
my_custom_plotter(axs[0], 'names', 'english', {'data':data,'color':'red','alpha':.3})
my_custom_plotter(axs[1], 'names', 'hindi', {'data':data,'color':'green','alpha':.3})
my_custom_plotter(axs[2], 'names', 'math', {'data':data, 'alpha':.3})
```
