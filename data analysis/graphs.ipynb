```python
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(5,5), layout='constrained')
axs[0,0].hist('petal_length', data=iris, bins=20, color='red')
axs[0,1].hist('petal_width', data=iris, bins=20, color='green')
axs[1,0].hist('sepal_length', data=iris, bins=20, color='blue')
axs[1,1].hist('sepal_width', data=iris, bins=20, color='pink')
axs[0,0].set_title('petal_length')
axs[0,1].set_title('petal_width')
axs[1,0].set_title('sepal_length')
axs[1,1].set_title('sepal_width')
plt.show()
```

```python
plt.boxplot('sepal_length', data=iris, notch=True, vert=False)
plt.show()
```

```python
plt.plot('sepal_length', data=iris, marker='o', color='purple', mfc='red')
```

```python
plt.bar(x= iris.index, height='sepal_length', data=iris, width=.2, color= 'purple')
```

```python
def convert_to_number(name):
    if name == 'setosa':
        return 1
    if name == 'verginica':
        return 2
    return 3

iris['spc'] = iris['species'].apply(convert_to_number)
iris
```

```python
plt.scatter(x='sepal_length',y='sepal_width',data=iris, c='spc', cmap='rainbow')
plt.colorbar()
plt.show()
```

```python
plt.scatter(x='petal_length',y='petal_width',data=iris, c='spc', cmap='rainbow')
```
