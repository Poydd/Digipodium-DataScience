```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt  # basic graphs
import seaborn as sns    # advanced graph
import plotly.express as px  #interactive graph
```

```python
titanic = sns.load_dataset('titanic')
titanic
```

```python
x = [1,2,3,4,5]
y = [5,6,7,8,9]
sizes = [30,40,50,60,70]
plt.scatter(x, y, s=sizes)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('bubble chart example')
plt.show()
```

```python
x = [1,2,3,4,5]
y = [5,6,7,8,9]

plt.plot(x, y, marker='s')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('line chart example')
plt.show()
```

```python
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(5,5), layout='constrained')
axs[0,0].hist('age', data=titanic, bins=20, color='red')
axs[0,1].hist('sex', data=titanic, bins=20, color='green')
axs[1,0].hist('fare', data=titanic, bins=20, color='blue')
axs[1,1].hist('who', data=titanic, bins=20, color='pink')
axs[0,0].set_title('age')
axs[0,1].set_title('sex')
axs[1,0].set_title('fare')
axs[1,1].set_title('who')
plt.show()
```

```python
plt.plot('age', data=titanic, marker='o', color='purple', mfc='red')
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
sns.lmplot(data=tips, x="total_bill", y="tip", col="time", hue='smoker')
```

```python
sns.displot(data=tips, x='total_bill', col='sex', kde='True')
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip')
```

```python
sns.catplot(data=tips, kind='swarm', x='sex', y='total_bill', hue='smoker')
```

```python
sns.catplot(data=tips, kind='violin', x='sex', y='total_bill', hue='smoker')
```

```python
sns.histplot(data=tips, x='sex', y='total_bill', hue='smoker')
```

```python
penguins = sns.load_dataset("penguins")
penguins
```

```python
sns.jointplot(data=penguins, x="flipper_length_mm", y="bill_length_mm", hue="species")
```

```python
sns.pairplot(data=penguins, hue='species')
```

```python
g = sns.PairGrid(penguins, hue="species", corner=True)
g.map_lower(sns.kdeplot, hue=None, levels=5, color=".2")
g.map_lower(sns.scatterplot, marker="+")
g.map_diag(sns.histplot, element="step", linewidth=0, kde=True)
g.add_legend(frameon=True)
g.legend.set_bbox_to_anchor((.61, .6))
```
