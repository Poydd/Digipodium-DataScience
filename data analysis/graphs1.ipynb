```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt  # basic graphs
import seaborn as sns    # advanced graph
import plotly.express as px  #interactive graph
```

```python
titanic = sns.load_dataset('titanic')
titanic
```

```python
x = [1,2,3,4,5]
y = [5,6,7,8,9]
sizes = [30,40,50,60,70]
plt.scatter(x, y, s=sizes)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('bubble chart example')
plt.show()
```

```python
x = [1,2,3,4,5]
y = [5,6,7,8,9]

plt.plot(x, y, marker='s')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('line chart example')
plt.show()
```

```python
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(5,5), layout='constrained')
axs[0,0].hist('age', data=titanic, bins=20, color='red')
axs[0,1].hist('sex', data=titanic, bins=20, color='green')
axs[1,0].hist('fare', data=titanic, bins=20, color='blue')
axs[1,1].hist('who', data=titanic, bins=20, color='pink')
axs[0,0].set_title('age')
axs[0,1].set_title('sex')
axs[1,0].set_title('fare')
axs[1,1].set_title('who')
plt.show()
```
